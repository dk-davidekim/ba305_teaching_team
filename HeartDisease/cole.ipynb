```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
```

```python
outcome = 'HeartDisease'
predictors = list(heart.columns)
predictors.remove(outcome)

X = heart[predictors]
y = heart[outcome]
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 305, stratify = y)

# After splitting the data, scale it so it performs well
scaler = preprocessing.StandardScaler()

scaler.fit(X_train)

X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# First: fully grown tree that will likely overfit

full_tree = DecisionTreeClassifier(random_state=305)
full_tree.fit(X_train_scaled, y_train)

y_pred_train = full_tree.predict(X_train_scaled)
y_pred_test = full_tree.predict(X_test_scaled)

train_score = accuracy_score(y_train, y_pred_train)
test_score = accuracy_score(y_test, y_pred_test)

print("Train score: ", train_score)
print("Test score: ", test_score)
```

```python
# Third: a tree with the best limitations using grid search
param_grid = {
    'max_depth': [10, 20, 30, 40],
    'min_samples_leaf': [20, 50, 100, 200],
    'min_impurity_decrease': [0, 0.0005, 0.001, 0.005, 0.01],
}


gridSearch = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring = 'accuracy')
gridSearch.fit(X_train_scaled, y_train)

print('Best accuracy: ', gridSearch.best_score_)
print('Parameter values to achieve best accuracy: ', gridSearch.best_params_)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from collections import Counter
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

```

```python
X, y = make_classification(n_classes=2, class_sep=2,
                           weights=[0.1, 0.9], n_informative=3, n_redundant=1, flip_y=0,
                           n_features=20, n_clusters_per_class=1, n_samples=1000, random_state=10)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print('Original training dataset shape %s' % Counter(y_train))
```

```python
sm = SMOTE(random_state=42)
X_train_res, y_train_res = sm.fit_resample(X_train, y_train)

print('Resampled training dataset shape %s' % Counter(y_train_res))
```

```python
tree = DecisionTreeClassifier(
      max_depth=10,
      min_samples_leaf = 20,
      min_impurity_decrease = 0.0005,
      criterion = 'gini')

tree.fit(X_train_res, y_train_res)
```

```python
y_pred = tree.predict(X_test)
score = f1_score(y_test, y_pred)
score
```
